### Filtro 1: Por objeto 

```python
# Exploramos los registros de nombre.
precios["NOMBRE"].value_counts()
```

```python
# Haremos un filtro que nos permita ver todas las instancias en las que el NOMBRE sea DIEGO ARMANDO LEON DURAN
filtro1=precios[precios["NOMBRE"] == "DIEGO ARMANDO LEON DURAN"]
filtro1
```

### Filtro 2: Por filas  

```python
# Supongamos que estamos interesados en mostrar la primera y la última instancia
filtro2=precios.iloc[ [0, 1425] , : ]  #Filas no consecutivas 
filtro2
```

### Filtro 3: Por columnas 

```python
#Filtraremos un dataframe con 4 columnas: NOMBRE, DESCR, PRECIO_UNITARIO Y COSTO_UNITARIO
filtro3=precios.iloc[ : , [2,6,7,8]] #Columnas no consecutivas
filtro3
```

### Filtro 4: Filas vs columnnas 

```python
# Utilizamos los valores de la columna 2 "NOMBRE" y los asignamos como índice 
df1= pd.read_csv('Detalle precios y productos fabricados 2022.csv', index_col=2)
df1
```

```python
# En este filtro utilizaremos la transformación anterior para mostrar los índices LETICIA RAMIREZ HERNANDEZ","AARON NUÑO CORDOVA
# Y su valor de la columna SUBTOTAL_PARTIDA
filtro4= df1.loc[["LETICIA RAMIREZ HERNANDEZ","AARON NUÑO CORDOVA"], ["SUBTOTAL_PARTIDA"]]
filtro4
```

### Filtro 5: Por cabecera 

```python
# Utilizamos la función head para mostrar los primeros 10 registros de nuestro dataframe 
filtro5 = precios.head(10)
```

### Filtros para variables cuantitativas 

### Filtro 6: Mayor qué 

```python
# Mostraremos las instancias en las que el COSTO_UNITARIO es mayor a 1000
# Para este filtro mostramos las instancias en las que los valores de CANT sean mayor que 10 
filtro6 = precios[precios["CANT"] > 10]
filtro6
```

### Filtro 7: Utilizando conectores lógicos 

```python
# Aquí filtraremos aquellas instancias en las que la columna CANT sea menor a 10 y la columna FECHA_DOC sea igual a 10/01/2022 00:00
filtro7=precios[(precios["CANT"] < 10) & (precios["FECHA_DOC"] == "10/01/2022 00:00")]
filtro7
```

### Filtro 8: Filtro o 

```python
# Aquí filtraremos aquellas instancias en las que la columna CANT sea menor a 5 o la columna FECHA_DOC sea igual a 10/01/2022 00:00
filtro8=precios[(precios["CANT"] < 5) | (precios["FECHA_DOC"] == "10/01/2022 00:00")]
filtro8
```

### Filtro 9: Filtro not 

```python
# # Aquí filtraremos aquellas instancias en las que la columna CANT no sean menor a 5 y
# en las que la columna FECHA_DOC no sea igual a 10/01/2022 00:00
filtro9=precios[~(precios["CANT"] >= 10) & ~(precios["FECHA_DOC"] == "23/12/2022 00:00")]
filtro9
```
