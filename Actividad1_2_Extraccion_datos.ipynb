```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Datos de Facturación.xlsx")
df
```

### Filtro 1: CVE_CLPV (valores: 1000 a 2000)

```python
# Filtramos los valores de las instancias de 1000 a 2000
df['CVE_CLPV'] = df['CVE_CLPV'].str.extract('(\d+)')
df['CVE_CLPV'] = df['CVE_CLPV'].dropna()
df = df.dropna(subset=['CVE_CLPV'])
df['CVE_CLPV'] = df['CVE_CLPV'].astype(int)

filtro1 =df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1
```

```python
df["CVE_CLPV"]
```

### Filtro 2: CVE_VEND (Todas las claves excepto 5 y 4)

```python
# Filtramos todos los valores para CVE_VEND a excepción de 5 y 4 
filtro2 = df[~df['CVE_VEND'].isin([4, 5])]
filtro2
```

### Filtro 3: FECHA_ENT(Fechas de 28/02/2022)

```python
# Primero convertimos la fecha 
df['FECHA_ENT'] = pd.to_timedelta(df['FECHA_ENT'], errors='coerce')
filtro3 = df[df["FECHA_ENT"] == "2022-02-28"]
filtro3 
# :(
```

### Filtro 4: CANT_TOT(Cantidades menores a 5951.7) o STATUS ("E")

```python
filtro4 = df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

### Sólo las columnas CVE_DOC, FECHA_ENT, FECHA_VEN, CANT_TOT

```python
filtro5 = df.iloc[ : , [0,6,7,9]]  #Columnas no consecutivas 
filtro5
```

### Sólo las filas de 7001-7099

```python
filtro6= df.iloc[7001:8000,: ]
filtro6
```

### Index = CVE_VEND(VALORES: 1 Y 2) (columna FECHAELAB)

```python
# Primero nos quedamos con los valores 1 y 2 de CVE_VEND
df_1 = df[df['CVE_VEND'].isin([1, 2])]
df_1 = df_1.set_index('CVE_VEND')
df_1

# Segunda parte 
filtro7 =df_1.loc[[1.0,2.0], ["FECHAELAB"]]
filtro7
```

### Generación de CSV

```python
filtro1.to_csv("FiltroA.csv")
filtro2.to_csv("FiltroB.csv")
filtro3.to_csv("FiltroC.csv")
filtro4.to_csv("FiltroD.csv")
filtro5.to_csv("FiltroE.csv")
filtro6.to_csv("FiltroF.csv")
filtro7.to_csv("FiltroG.csv")
```
